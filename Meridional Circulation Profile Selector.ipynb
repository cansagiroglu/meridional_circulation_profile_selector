```python
###Meridional Circulation Profile Selector, CS,2018###
import math
import matplotlib.pyplot as plt
import numpy as np
profile=input("Which meridional circulation profile will be used? (1)for Dikpati_Charbonneau_99,(2)Dikpati_et.al_2004")
art=input("What should be the increment for the radius grid? 0.01, 0.02, 0.05, 0.1?")
tart=input("What should be the increment for the theta grid? 1.0, 2.0, 3.0, 5.0, 10.0")
u0=input("What is u0? Please write a value between 1.0-200.0")

urr=[] ##ur component dependent on both theta and r
utt=[] ##ut component dependent on both theta and r

####Creating r and theta points for a quarter circle
RR=1.0
r0=0.68*RR
rr=[r0] ##radii array for calculation, starting from r0
t=0.0
tt=[t] ## degree degree array for calculation
tr=[t] ##degree radian array for plot
basson=RR-r0
if art==0.01:
    art=basson/(basson*100.0)
    lengthofr=int(basson*100.0)+1
elif art==0.02:
    art=basson/(basson*50.0)
    lengthofr=int(basson*50.0)+1
elif art==0.05:
    art=basson/(basson*20.0)
    lengthofr=int(basson*20.0)
elif art==0.1:
    art=basson/(basson*10.0)
    lengthofr=int(basson*10.0)
lengthoft=int(round(90.0/tart))
tplot=[]
ttplot=[]
for i in range (lengthoft):
    radian=math.radians(t)
    tr.append(radian)
    t=t+tart
    tt.append(t)
lengtht=len(tt)
tplot=(tr*(lengthofr+1)) ##repeating theta for r grid point see DebugKey1 for proof.
ttplot=(tt*(lengthofr+1))
tplotlen=len(tplot)

rplot0=[r0]
rplot=rplot0*(lengtht)##radii array for plotting
for k in range (lengthofr):
    r0=r0+art
    round1=round(r0,2)
    a=[round1]
    rr.append(round1)
    x=a*lengtht
    rplot=rplot+x ###For proof see DebugKey1.
rlength=len(rr)
r0=0.68*RR


##Dikpati_etal_2004
if profile==2:
    m=1.5
    b=3.0
    q=1.5
    u0=-1.0*u0
    xib=(RR/r0)-1.0 
    c1=(2*m+1)*(m+b)/((m+1)*b)*((xib)**(-m))
    c2=(2*m+b+1)*m/((m+1)*b)*((xib)**(-(m+b)))
    k=0
    j=0
    p=0
    for j in range (lengthofr+1):
        for p in range(lengtht):
            xi=RR/rr[j]-1.0
            urtop=u0*((RR/rr[j]))*((-1.0/(m+1.0))+(c1/(2.0*m+1.0))*xi**(m)-(c2/(2.0*m+b+1.0))*(xi**(m+b)))*xi*((math.sin(math.radians(tt[p])))**q)*((q+2.0)*((math.cos(math.radians(tt[p])))**2.0)-((math.sin(math.radians(tt[p])))**2.0))
            uttop=u0*((RR/rr[j])**3.0)*(-1.0+c1*((xi)**(m))-c2*((xi)**(m+b)))*((math.sin(math.radians(tt[p])))**(q+1.0))*(math.cos(math.radians(tt[p])))
            urr.append(urtop)
            utt.append(uttop)
#            print "k=",k,"j=",j,"p=",p,"r=",rr[j],"derece=",tt[p],"ur=",urr[k],"ut=",utt[k]#, "urtop=",urtop,"uttop=",uttop ##Debugkey1
            k=k+1
    utir=[]
    g=0
    u0=-1.0*u0
    for g in range (lengthofr+1):
        xi=RR/rr[g]-1.0 
        utirr=u0*((RR/rr[g])**3.0)*(-1.0+c1*((xi)**(m))-c2*((xi)**(m+b)))*((math.sin(math.radians(45)))**(q+1.0))*(math.cos(math.radians(45)))
        utir.append(utirr)
    frrr=plt.figure()
    if (u0>=1.0 and u0<=5.0):
        sss=0.2
    elif (u0>5.0 and u0<=20.0):
        sss=u0/u0
    elif (u0>20.0 and u0<=50.0):
        sss=u0/10
    elif (u0>50.0 and u0<=100.0):
        sss=5.0
    elif (u0>100.0 and u0<=200.0):
        sss=20.0
    minutir=min(utir)
    maxutir=max(utir)
    plt.plot(rr,utir,ls="solid", color='black')
    plt.margins(0.01)
    plt.axvline(x=0.7,lw=1,ls="dashed", color='blue')
    plt.xticks(np.arange(0.68,1.01,step=0.02),fontsize=20)
    plt.yticks(fontsize=20)
    #plt.yticks(np.arange(minutir-(sss+(sss/4.0)), maxutir+(sss+(sss/4.0)), step=sss))
#    plt.yticks(np.arange(minutir-(sss+(sss/4.0)), maxutir+(sss+(sss/4.0)), step=sss))
    plt.xlabel('r/R(Sun)',fontsize=20)
    plt.ylabel('u theta(r,theta=45degrees)',fontsize=20)
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    ru0=round(u0,2)
    fig.savefig('ut@45_u=%s.png'%(ru0), dpi=500)
    
##Dikpati_Charbonneau_99
elif profile==1: 
    m=0.5
    b=0.25
    q=0.0
    u0=-1.0*u0 
    xib=(RR/r0)-1.0 
    c1=(2*m+1)*(m+b)/((m+1)*b)*((xib)**(-m))
    c2=(2*m+b+1)*m/((m+1)*b)*((xib)**(-(m+b)))
    k=0
    j=0
    p=0
    for j in range (lengthofr+1):
        for p in range(lengtht):
            xi=RR/rr[j]-1.0
            urtop=u0*((RR/rr[j])**2.0)*((-1.0/(m+1.0))+(c1/(2.0*m+1.0))*xi**(m)-(c2/(2.0*m+b+1.0))*(xi**(m+b)))*xi*((math.sin(math.radians(tt[p])))**q)*((q+2.0)*((math.cos(math.radians(tt[p])))**2.0)-((math.sin(math.radians(tt[p])))**2.0))
            uttop=u0*((RR/rr[j])**3.0)*(-1.0+c1*((xi)**(m))-c2*((xi)**(m+b)))*((math.sin(math.radians(tt[p])))**(q+1.0))*(math.cos(math.radians(tt[p])))
            urr.append(urtop)
            utt.append(uttop)
#            print "k=",k,"j=",j,"p=",p,"r=",rr[j],"derece=",tt[p],"ur=",urr[k],"ut=",utt[k]#, "urtop=",urtop,"uttop=",uttop ##Debugkey1
            k=k+1
    utir=[]
    g=0
    u0=-1.0*u0
    for g in range (lengthofr+1):
        xi=RR/rr[g]-1.0 
        utirr=u0*((RR/rr[g])**3.0)*(-1.0+c1*((xi)**(m))-c2*((xi)**(m+b)))*((math.sin(math.radians(45)))**(q+1.0))*(math.cos(math.radians(45)))
        utir.append(utirr)
    frrr=plt.figure()
    plt.plot(rr,utir,ls="solid", color='black')
    minutir=min(utir)
    maxutir=max(utir)
    if (u0>=1.0 and u0<=5.0):
        sss=0.2
    elif (u0>5.0 and u0<=20.0):
        sss=u0/u0
    elif (u0>20.0 and u0<=50.0):
        sss=u0/10
    elif (u0>50.0 and u0<=100.0):
        sss=5.0
    elif (u0>100.0 and u0<=200.0):
        sss=20.0
    plt.margins(0.0)
    plt.axvline(x=0.7,lw=1,ls="dashed", color='blue')
    plt.xticks(np.arange(0.68,1.01,step=0.02),fontsize=18)
    plt.yticks(np.arange(minutir, maxutir+(sss+(sss/2.0)), step=sss),fontsize=20)
    plt.xlabel('r/R(Sun)',fontsize=20)
    plt.ylabel('u theta(r,theta=45degrees)',fontsize=20)
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    ru0=round(u0,2)
    fig.savefig('ut@45_u=%s.png'%(ru0), dpi=500)


####plotting the quarter circle
##Cylindirical to Cartesian Conversion because symmetric in phi.
xr_dat=[]
yt_dat=[]
s=0
for s in range (tplotlen): 
    x_datt=urr[s]*math.cos(tplot[s])-utt[s]*math.sin(tplot[s])
    xr_dat.append(x_datt)
    y_datt=urr[s]*math.sin(tplot[s])+utt[s]*math.cos(tplot[s])
    yt_dat.append(y_datt)


f = plt.figure()
ax = f.add_subplot(111, polar=True,zorder=2)
ax.quiver(tplot, rplot, xr_dat, yt_dat)

##setting the plotting preferences
ax.set_thetamin(0)
ax.set_thetamax(90)
ax.set_rmin(0)
ax.set_rmax(1)
ax.set_rticks([0.6,0.7, 0.8, 0.9,1.0])
ax.set_thetamin(0)
ax.set_thetamax(90)
ax.set_thetagrids([15,30,45,60,75,90])

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

rart=round(art,2)
rtart=round(tart,2)

fig.savefig('CirculationProfile_p=%s_incr=%s_int=%s.png'%(profile,rart,rtart), dpi=500)
```
